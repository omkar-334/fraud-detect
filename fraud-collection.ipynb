```python
import json

from data import add_info

with open("dataset\\fraud-apps.json", encoding="utf-8") as f:
    fraud = json.load(f)
```

```python
expanded = []
for i in range(len(fraud)):
    try:
        print(f"-----{i}-----")
        old = fraud[i]
        new = add_info(old)
        expanded.append(new)
    except Exception as e:
        print(f"XXXX {i} XXXX - {e}")
```

```python
with open("dataset/fraud-expanded.json", "wb") as f:
    f.write(json.dumps(expanded, indent=4).encode("utf-8"))
    print("File written successfully")
```
